```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
nulos_por_coluna = vendas.isnull().sum()
print(nulos_por_coluna)
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'] = vendas['estado'].fillna('MS')
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'] = vendas['Preço'].fillna(media_preco)
media_preco
```

```python
vendas.info()
```

```python
mediana_preco = vendas['Preço'].median()
mediana_preco
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preço > Preço_com_frete')
vendas_corretas = vendas.query('Preço < Preço_com_frete')
```

```python
vendas_corretas.describe()
```

```python
vendas_corretas.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_corretas.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(), 2).reset_index()
```

```python
vendas_corretas['mes'] = vendas_corretas['Data'].dt.month
vendas_corretas.head()
```

```python
cliente = pd.read_excel('/content/varejo.xlsx')
cliente.head()
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
vendas_corretas.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
nulos_por_coluna = cliente.isnull().sum()
print(nulos_por_coluna)
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_corretas.merge(cliente, on='cliente_Log', how='left')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda') ['renda'].agg('mean').sort_values(ascending=False).reset_index(), 2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira') ['idade'].agg('mean').sort_values(ascending=False).reset_index(), 2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')
plt.title('Média de Idade por Bandeira', loc='center', fontsize=18, fontweight='bold', color = 'black')
plt.xlabel('Bandeira')
plt.ylabel('Média de Idade');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'pink')
plt.title('Média de renda por canal de venda', loc='center', fontsize=18, fontweight='bold', color = 'black')
plt.xlabel('Canal de venda')
plt.ylabel('Média de renda');
```

```python
venda_por_data = vendas_corretas.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
venda_por_data
```

```python
venda_por_data = venda_por_data.sort_values(by="Data")
venda_por_data['Data'] = pd.to_datetime(venda_por_data['Data'])
```

```python
fig = plt.figure(figsize=(20,12))


plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x,y, color = "green", linewidth = "2")
plt.show()
```

```python
agg_dept_preco = round(vendas_corretas.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(30,20))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"], color="green")

plt.ylabel("Preço médio")

plt.title("Preço médio por departamento", loc="center", fontsize=18, fontweight="bold", color="black");
```
